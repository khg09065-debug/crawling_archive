```python
import pandas as pd
from sqlalchemy import create_engine

# 1. MySQL 연결 설정 (본인의 DB 접속 정보로 수정하세요)
DB_USER = "root"          # MySQL 사용자 아이디
DB_PASSWORD = "1234"  # MySQL 비밀번호
DB_HOST = "localhost"     # 호스트 주소 (예: localhost 또는 IP 주소)
DB_PORT = "3306"          # 포트 번호 (기본값 3306)
DB_NAME = "coffee_store"  # 데이터베이스명

# 데이터베이스 연결 엔진 생성
engine = create_engine(f"mysql+pymysql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}")

def print_brand_counts():
    try:
        # 2. SQL 쿼리 작성 (브랜드명별로 그룹화하여 개수 계산)
        query = """
        SELECT `브랜드명`, COUNT(*) AS `개수`
        FROM `coffee_chain`
        GROUP BY `브랜드명`
        ORDER BY `개수` DESC;
        """
        
        # 3. 쿼리 실행 후 Pandas DataFrame으로 가져오기
        df = pd.read_sql(query, con=engine)
        
        if df.empty:
            print("테이블에 데이터가 존재하지 않습니다.")
            return

        print(f"찾은 브랜드 개수: {len(df)}개\n")
        
        # 4. 결과 출력
        for index, row in df.iterrows():
            print(f"- {row['브랜드명']}: {row['개수']}개")
            
        print("-" * 30)
        print(f"전체 데이터 행 수 합계: {df['개수'].sum()}개")
        
    except Exception as e:
        print(f"데이터베이스 연결 또는 쿼리 실행 실패: {e}")

# 함수 실행
if __name__ == "__main__":
    print_brand_counts()
```
