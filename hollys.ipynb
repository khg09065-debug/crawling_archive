```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time

# 1. 브라우저 설정
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

url = "https://m.hollys.co.kr/store/korStore2.do"
driver.get(url)
wait = WebDriverWait(driver, 10)

def collect_hollys():
    results = []
    
    # 2. '더보기' 버튼이 없어질 때까지 클릭 (클래스명 btn_wrap shop_more 반영)
    print("--- 매장 목록 확장 시작 ---")
    while True:
        try:
            # btn_wrap.shop_more 클래스 안에 있는 실제 클릭 버튼(a 태그 등)을 찾음
            more_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".btn_wrap.shop_more a")))
            driver.execute_script("arguments[0].click();", more_button) # JS 클릭으로 확실하게 실행
            time.sleep(1.5) # 로딩 대기
            print("더보기 클릭 중...")
        except:
            print("더보기 완료: 모든 매장을 로드했습니다.")
            break

    # 3. 매장 목록 가져오기
    stores = driver.find_elements(By.CSS_SELECTOR, ".searchResult .store_area")
    total_stores = len(stores)
    print(f"총 {total_stores}개의 매장을 찾았습니다.")

    # 4~6. 반복 작업 수행
    for i in range(total_stores):
        start_time = time.time()
        
        try:
            target_store = driver.find_elements(By.CSS_SELECTOR, ".searchResult .store_area")[i]
            
            # 매장명 추출
            soup_list = BeautifulSoup(target_store.get_attribute('innerHTML'), 'html.parser')
            store_name = soup_list.get_text(strip=True)
            
            # 상세 페이지 진입
            target_store.click()
            time.sleep(1.5)
            
            # 5. 상세창에서 '주소' 텍스트를 가진 항목을 찾아 데이터 추출
            detail_soup = BeautifulSoup(driver.page_source, 'html.parser')
            address = ""
            
            li_t_tags = detail_soup.find_all("span", class_="li_t")
            for tag in li_t_tags:
                if "주소" in tag.get_text():
                    address = tag.find_next_sibling("span", class_="li_con").get_text(strip=True)
                    break
            
            results.append({"매장명": store_name, "주소": address})
            
            # 6. 목록으로 돌아가기
            back_btn = wait.until(EC.element_to_be_clickable((By.CLASS_NAME, "btn")))
            back_btn.click()
            time.sleep(1)
            
            end_time = time.time()
            print(f"[{i+1}/{total_stores}] {store_name} 완료 ({end_time - start_time:.2f}초)")

        except Exception as e:
            print(f"{i+1}번째 매장 수집 중 오류 발생: {e}")
            continue

    # 7. 데이터 저장
    df = pd.DataFrame(results)
    df.to_csv("hollys.csv", index=False, encoding="utf-8-sig")
    print("--- 수집 완료 및 hollys.csv 저장 성공 ---")

try:
    collect_hollys()
finally:
    driver.quit()
```
