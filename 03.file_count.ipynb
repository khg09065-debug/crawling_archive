```python
import pandas as pd
import glob
import os

def count_total_csv_rows():
    # 1. 현재 경로에서 모든 csv 파일 목록을 가져옵니다.
    csv_files = glob.glob('*.csv')
    
    if not csv_files:
        print("현재 폴더에 CSV 파일이 없습니다.")
        return

    total_rows = 0
    print(f"찾은 CSV 파일 개수: {len(csv_files)}개\n")

    # 2. 각 파일별로 행 수를 확인합니다.
    for file in csv_files:
        try:
            # 파일을 읽어 행 수를 계산 (헤더 제외)
            df = pd.read_csv(file)
            row_count = len(df)
            total_rows += row_count
            print(f"- {file}: {row_count}개")
        except Exception as e:
            print(f"- {file}: 읽기 실패 ({e})")

    # 3. 전체 합계 출력
    print("-" * 30)
    print(f"전체 데이터 행 수 합계: {total_rows}개")

# 함수 실행
if __name__ == "__main__":
    count_total_csv_rows()
```
