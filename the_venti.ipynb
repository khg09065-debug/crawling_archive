```python
import time
import pandas as pd
import requests
from bs4 import BeautifulSoup

def scrape_the_venti():
    all_data = []
    page_num = 1
    
    print("더벤티 매장 정보 수집을 시작합니다... (한글 인코딩 처리 적용)")
    print("-" * 50)

    # 세션을 사용하면 연결이 더 안정적입니다.
    session = requests.Session()
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'}

    while True:
        target_url = f"https://www.theventi.co.kr/new2022/store/index.html?page={page_num}&plist=&find_field=&find_word=&find_state=&find_ordby=&conf=&find_mode=&city=&county="
        
        try:
            response = session.get(target_url, headers=headers)
            
            # [수정 포인트 1] 응답 인코딩을 명시적으로 UTF-8로 설정하여 한글 깨짐 방지
            response.encoding = 'utf-8' 
            
            soup = BeautifulSoup(response.text, 'html.parser')
            
            menu_body = soup.select_one('.menu_body')
            stores = menu_body.select('.tr') if menu_body else []

            if not stores:
                print(f"\n[알림] {page_num}페이지에 데이터가 없습니다. 수집을 종료합니다.")
                break

            for store in stores:
                # 매점명 추출 (tit_b_mc 제외)
                col_tit = store.select_one('.col_tit')
                if col_tit:
                    # 원본을 건드리지 않기 위해 복사해서 처리
                    name_tag = col_tit.find(text=True, recursive=False)
                    store_name = name_tag.strip() if name_tag else col_tit.get_text(strip=True)
                else:
                    store_name = "N/A"

                # 주소 추출
                address_div = store.select_one('.col_info .address')
                address = address_div.get_text(strip=True) if address_div else "N/A"

                all_data.append({'매점명': store_name, '주소': address})

            print(f"[{page_num} 페이지 완료] 누적 데이터: {len(all_data)}개")
            
            page_num += 1
            time.sleep(0.3)

        except Exception as e:
            print(f"오류 발생: {e}")
            break

    # [수정 포인트 2] CSV 저장 시 utf-8-sig를 사용하여 엑셀 한글 깨짐 방지
    if all_data:
        df = pd.DataFrame(all_data)
        df.to_csv('the_venti.csv', index=False, encoding='utf-8-sig')
        print("-" * 50)
        print(f"최종 저장 완료: the_venti.csv (총 {len(df)}개 매장)")

if __name__ == "__main__":
    scrape_the_venti()
```
